```python
params = {
    'limit': '5',
    'include': 'comments,contribute_info',
    'sort': 'score|desc,id|desc,stars|all',
    'page': '1',
    'spid': '69247930',
    'product_id': '68793627',
    'seller_id': '97516'
}
```

```python
Tiki_comment = pd.DataFrame()
Comment = []
Id = []
count_id = 0
page_num = 0
```

```python
response = requests.get('https://tiki.vn/api/v2/reviews', headers=headers, params=params)
```

```python
while len(Comment) < 400:
    page_num += 1
    params['page'] = page_num
    response = requests.get('https://tiki.vn/api/v2/reviews', headers=headers, params=params)
    print('Crawl comment page {} success!!!'.format(page_num))
    for cmt in response.json().get('data'):
        cmts = cmt.get('content')
        if len(cmts.split()) > 2:
            Comment.append(cmt.get('content'))
            count_id += 1
            Id.append(count_id)
```

```python
Tiki_comment['Id'] = Id
Tiki_comment['Comment'] = Comment
```
